```python
data =[0]
def rules(n):
    nstr = str(n)
    lstr = int(len(str(nstr)))
    if n == 0:
        return [1]
    elif len(nstr) % 2 == 0:
        return [int(nstr[0:int(lstr / 2)]),int(nstr[int(lstr / 2):lstr])]
    else:
        return [n * 2024]

looper = data
for i in range(15):
    print(i)
    loop_res = []
    for item in looper:
        res=rules(item)
        for subres in res:
            loop_res.append(subres)
    looper = loop_res
    print(looper)
    print(len(looper))
```
