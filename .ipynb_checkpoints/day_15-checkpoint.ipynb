```python
with open('Inputs/day_15.txt','r') as file:
    lines = file.readlines()
    
data_map = []
data_move = []
for line in lines:
    row = list(line.replace('\n',''))
    if len(row) == 0:
        print(row)
    elif row[0] == '#':
        data_map.append(row)
    elif row[0] in ['v','<','>','^']:
        for item in row:
            data_move.append(item)
        
for x in range (len(data_map[0])):
    for y in range (len(data_map)):
        if data_map[y][x] == '@':
            x_pos = x
            y_pos = y
```

```python
def move_stuff(x,y,direction,set_val):    
    if direction == '^':
        y=y-1
    elif direction == '>':
        x=x+1
    elif direction == 'v':
        y=y+1
    elif direction == '<':
        x=x-1
    if data_map[y][x] == '#':
        changes.append([0,0,''])
    elif data_map[y][x] == 'O':
        enum = move_stuff(x,y,direction,'O')
        for row in enum:
            changes.append(row)
        changes.append([x,y,set_val])
    elif data_map[y][x] == '.':
        changes.append([x,y,set_val])
    else:
        print('Error')
        print(data_map[y][x])
        print(x)
        print(y)
    return changes
```
